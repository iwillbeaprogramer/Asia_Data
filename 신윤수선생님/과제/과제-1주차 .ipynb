# 5.  리스트와 set
다음과 같은 리스트가 선언되어있다.
```
>> lang1 = ["C", "C++", "JAVA", "Perl", "R", "Python"]
>> lang2 = ["Python", "Go", "C#", "JAVA"]
```

```lang1```과 ```lang2``` 의 원소를 모두 갖고 있는 변수 ```lang3```를 만들어라.   
단 중복은 제거하라. (배치 순서는 상관하지 않음.)


```
실행 예:
>> langs
['C', 'C++', 'JAVA', 'Python', 'Go', 'C#', 'R', 'Perl']
```

````
lang1 = ["C", "C++", "JAVA", "Perl", "R", "Python"]
lang2 = ["Python", "Go", "C#", "JAVA"]

lang3 = (lang1 + lang2)

lang_set = set(lang3)
lang_set
````

## 3. 피보나치수열
f(n)은 피보나치 수열의 n번째 숫자이다. 사용자로부터 n을 입력받아 피보나치 수의 n번째 항을 출력한다.  

피보나치 수열: f(n) = f(n-1) + f(n-2) 이며   
f(1) = 1  
f(2) = 1이라 하자  

**[출력예시]**   
f(3) = 2  
f(4) = 3  
f(5) = 5

```python
# 카페 정보를 딕셔너리로 가져오는 함수
def bring_cafe_info(query):
    map_base_url = 'https://map.naver.com/v5/api/search'
    params = {'caller':'pcweb',
              'query': "종로구 "+query+' 카페',
              'type':'all',
              'searchCoord':'126.88522338867189;37.47744648556929',
              'page':0,
              'displayCount':20,
              'isPlaceRecommendationReplace':'true',
              'lang':'ko'}
    
    headers = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/84.0.4147.105 Safari/537.36'}
    
    is_ended = False
    cafe_dict= {}
    while not is_ended:
        params['page']+=1
        resp = requests.get(map_base_url,  headers=headers, params=params)
        cafes = json.loads(resp.text)
        
        cafe_result_dict= cafes.get('result')
        if not cafe_result_dict:
            break
        place = cafe_result_dict.get('place')
        if not place:
            break
        cafe_info_list = place['list']
        if not cafe_info_list:
            break
        

        for cafe in cafe_info_list:
            a = cafe['roadAddress']
            m= cafe["menuInfo"]
            if a[6:9] == '종로구':
                n = cafe['name']
                ct = cafe["category"]
                rc = cafe['reviewCount']
                cafe_dict[n] = [ct, m, a, rc]
                
            
    return cafe_dict
```

```python
# 숫자 내에 ~와 같은 기호가 있을 때 분리하고 정수형으로 반환하는 함수
def get_price(price_str):
    token = '~!@#$%^&*()_+'
    for i in token:
        # 기호 기준으로 나눠서 앞 가격만 가져오기 
        price_str = price_str.split(i)[0]
    return int(price_str)
```

```python
def split_menu_and_price(cafe_dict):
    result_list= []
    
    for keys in cafe_dict:
        cafe_info_dict= {}
        menu_dict= {}
        # m: 메뉴 자체가 없는 경우
        m = cafe_dict[keys][1]
        # 카페 주소
        a= cafe_dict[keys][2]
        #  메뉴가 없으면 넘어가기 
        if m == None:
            continue
        else:
            # 메뉴 문자열에 들어있는 '|' 제거
            menus = m.split('|')
            menu_and_price = []
       
            for menu in menus:
                menu = menu.strip()
                # 가격과 메뉴 분리
                menu_detail = menu.split(' ')
                menu_and_price.append(menu_detail)
            
            # 가격 내에 있는 ',' 지우기
            for price in menu_and_price:
                num_price = price[-1]
                num = num_price.find(',')
                name_list = price[:-1]
                name = ' '.join(name_list)
                
                if num > 0:
                    comma = num_price.split(num_price[num])
                    real_price= ''.join(comma)
                    # 가격 내에 또 기호가 존재하는 경우  get_price함수 이용
                    int_price = get_price(real_price)
                    menu_dict[name] = int_price
                
                # 가격이 없을 경우 (None)   
                else:
                    continue
            
            cafe_info_dict["name"]= keys
            cafe_info_dict['address']= a
            cafe_info_dict['menu']= menu_dict
            
            
            result_list.append(cafe_info_dict)
            
    return result_list
            
```

```python
# 아메리카노가 있는 카페의 이름, 주소, 아메리카노 가격을 알려주는 함수
def check_only_americano(cafe_list):
    i = 50000
    final_result_dict ={}
    for cafes in cafe_list:
    # 카페 메뉴 정보 리스트 내에 있는 딕셔너리를 부르는 것
        menu = cafes['menu']
        name = cafes['name']
        add = cafes['address']
        # 메뉴 중에서 아메리카노가 있는지 확인 
        if '아메리카노' in menu:
            if menu['아메리카노'] != None:
            # 아메리카노 가격 중에서 최저가인 카페 정보를 딕셔너리에 추가
                if menu['아메리카노'] < i:
                    i = menu['아메리카노']
                    final_result_dict = {'이름': name, 
                                         '주소': add, 
                                         '아메리카노': menu['아메리카노']}
        else:
            pass
    
    return final_result_dict
```

```python
# 현재 웹크롤링 상태를 확인하는 함수 
def get_state():
    data = None
    # 현재 상태를 기록한 제이슨 파일을 읽어와라
    try:
        with open('state.json', 'r') as f:
            data = f.read()
            data = json.loads(data)
    except:
        # 파일이 존재하지 않을 경우 도로명주소 엑셀을 불러와라 
        df = pd.read_excel('road_names.xlsx')
        
        # 도로명 주소들을 딕셔너리에 키워드로 넣기 
        return {
            'all': df.iloc[:, 0].values,
            'done': [],
            'wait': df.iloc[:, 0].values
        }
    return data
```

```python
# 현재까지 웹크롤링한 내용을 저장하는 함수 
def save_state(state):
    with open("state.json", 'w' ) as f:
        f.write(json.dumps(state))
    return True
```

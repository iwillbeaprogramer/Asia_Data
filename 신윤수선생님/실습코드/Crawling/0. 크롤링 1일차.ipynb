# 방탄소년단 뉴스 리스트

- title
- url

- description(+알파)

````python
url = 'https://search.naver.com/search.naver?sm=tab_hty.top&where=news&query=%EB%B0%A9%ED%83%84%EC%86%8C%EB%85%84%EB%8B%A8&oquery=%EB%B0%A9%ED%83%84%EC%86%8C%EB%85%84%EB%8B%A8&tqi=Uy42Ywp0J1ZssLzVHVhssssssuR-089801'
resp = requests.get(url)
resp
````

### 방법1

````python
soup = BeautifulSoup(resp.text)
ul_tags = soup.find('ul', class_='type01')
news_li = ul_tags.find_all('li', recursive=False)
print(len(news_li))
news_li
````

````python
result_list = []
for news in news_li:
    title_tag = news.find('a', class_='_sp_each_url')
    title = title_tag.text
    url = title_tag.get('href')
    
    desc = news.find_all('dd')
    desc = desc[1].text
    result_list.append({
        'title': title,
        'url': url,
        'desc': desc
    })
print(len(result_list))
result_list
````

### 방법2

````python
print(len(soup.select('ul.type01>li>dl>dt>a')))
titles = soup.select('ul.type01>li>dl>dt>a')
````

````python
#css selector
print(len(soup.select('ul.type01>li>dl>dd:nth-child(3)')))
descs = soup.select('ul.type01>li>dl>dd:nth-child(3)')
````

````python
result_list = []
for a_tag, desc_tag in zip(titles, descs):
    result_list.append({
        'title': a_tag.text,
        'url': a_tag.get('href'),
        'desc': desc_tag.text
    })
````

````python
print(len(result_list))
result_list
````

## 블로그 리스트 5페이지까지..

#### Basic

````python
url = 'https://search.naver.com/search.naver'

page_num = int(input("Page Number?"))
blog_list = []
for i in range(page_num):
    params = {
        'where': 'post',
        'query': "방탄소년단",
        "start": i*10 + 1
    }

    resp = requests.get(url, params=params)
    resp

    soup = BeautifulSoup(resp.text)

    blog_list_tags = soup.find_all('li', class_='sh_blog_top')

    for blog in blog_list_tags:
        a_tag = blog.find('a', class_='sh_blog_title')
        blog_url = a_tag.get('href')

        title = a_tag.text
        desc = blog.find_all('dd')[1].text

        blog_list.append({
            'url': blog_url,
            'title': title,
            'desc': desc
        })
blog_list

````

## 블로그 크롤링 함수(페이지,쿼리) 

````python
def crawl_blog(query, page_num=2):
    url = 'https://search.naver.com/search.naver'
    blog_list = []
    
    for i in range(page_num):
        params = {
            'where': 'post',
            'query': query,
            "start": i*10 + 1
        }

        resp = requests.get(url, params=params)
        resp

        soup = BeautifulSoup(resp.text)

        blog_list_tags = soup.find_all('li', class_='sh_blog_top')

        for blog in blog_list_tags:
            a_tag = blog.find('a', class_='sh_blog_title')
            blog_url = a_tag.get('href')

            title = a_tag.text
            desc = blog.find_all('dd')[1].text

            blog_list.append({
                'url': blog_url,
                'title': title,
                'desc': desc
            })
    return blog_list

````

````python
blog_search_list = crawl_blog("트와이스", 3)
print(len(blog_search_list))
blog_search_list
````

## 뉴스 상세페이지

```
[
    {
        'title': ~~,
        'url': ~~,
        'content': ~~
    }
    ...
]
```

````python
base_url = 'https://search.naver.com/search.naver'
params = {
    'where': 'news',
    'query': '방탄소년단'
}
resp = requests.get(base_url, params=params)
resp
````

````python
soup = BeautifulSoup(resp.text)
a_tags = soup.select('ul.type01>li>dl>dt>a._sp_each_title')
# a_tags = 
len(a_tags)
````

````python
result_list = []
for a_tag in a_tags:
    title = a_tag.text
    news_url = a_tag.get('href')
    
    detail_resp = requests.get(news_url)
    
    content = detail_resp.text
    result_list.append({
        'title':title,
        'url': news_url,
        'content': content
    })
````

```python
result_list = []
for news_item in news_list_tags:
    dt_tags = news_item.find_all('dt')
    if len(dt_tags) == 1:
        # 이미지X
        title_tag = dt_tags[0]
    else:
        # 이미지 O
        title_tag = dt_tags[1]
    result_list.append({
        'url': title_tag.find('a').get('href'),
        'title': title_tag.text.strip()
    })
result_list
```

### 함수화

```python
def get_naver_news_list(sid1, sid2=None, date=None):
    base_url = 'https://news.naver.com/main/list.nhn'
    params = {
        'mode':'LS2D',
        'mid':'shm',
        'sid1': sid1,
        'sid2': sid2,
        'date': date,
        'page': 0
    }
    is_ended = False
    result_list = []
    
    while not is_ended:
        params['page'] += 1
        resp = requests.get(base_url, params=params)
        soup = BeautifulSoup(resp.text, 'lxml')
        
        news_list_tags = soup.select('div.list_body > ul > li')
        
        num_dup = 0
        for news_item in news_list_tags:
            dt_tags = news_item.find_all('dt')
            if len(dt_tags) == 1:
                # 이미지X
                title_tag = dt_tags[0]
            else:
                # 이미지 O
                title_tag = dt_tags[1]
            
            news = {
                'url': title_tag.find('a').get('href'),
                'title': title_tag.text.strip()
            }
            if num_dup >=3:
                is_ended = True
                break
            
            if news in result_list:
                num_dup +=1
                continue
                
            result_list.append(news)
            
            
        if len(news_list_tags) < 20:
            is_ended = True
        
    return result_list
            
        
    
    
```

```python
news_list = get_naver_news_list(105, 731, '20200802')
len(news_list)
```

### 본문 가져오기
